```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# مدل را ایجاد کنید (با فرض کد مدل قبلی)
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(28, 28, 3)),
    tf.keras.layers.Conv2D(filters=16, kernel_size=(1, 1), activation='relu'),
])

# نمونه‌ای از تصویر ورودی
input_image = np.random.rand(1, 28, 28, 3)

# محاسبه خروجی لایه
layer_output = model.predict(input_image)

# نمایش تصویر ورودی
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title('Input Image')
plt.imshow(input_image[0].astype('uint8'))  # نمایش تصویر ورودی

```

```python
# نمایش خروجی لایه
plt.subplot(1, 2, 2)
plt.title('Layer Output')

for i in range(layer_output.shape[-1]):
    plt.subplot(4, 4, i+1)
    plt.imshow(layer_output[0, :, :, i], cmap='gray')  # نمایش هر کانال به صورت تصویر خاکستری
plt.show()
```
